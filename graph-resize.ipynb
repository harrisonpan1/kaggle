```python
import cv2
from tqdm import tqdm_notebook as tqdm
from glob import glob
```

```python
filelist = glob('/kaggle/input/aptos2019-blindness-detection/train_images/*')
```

```python
len(filelist)
```

```python
size = 512
path = 'kaggle/output/resized_train_images'
for filename in tqdm(filelist):
    image = cv2.imread(filename)
    res = cv2.resize(image, (size, size), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(path + filename.split('/')[-1], res)
```
